```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Pandas is for data manipulation and analysis
2. Numpy is for numerical operations
3. Matplotlib.pyplot and seaborn are used for data visualisation


Create Data Frame


```python
dataframe = pd.read_csv('Zomato data .csv')
print(dataframe)
```

User defined function

```python
dataframe
```

```python
def handleRate(value):
  value = str(value).split('/') #removes the /5 part from the rating
  value = value[0];
  return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()  ## to check whetherb there are any missing values or any null values in the dataset
```

object is string type data

Type of Restaurant

```python
dataframe.head()  # First 5 data it shows head() function
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of Restaurant")
## sns.countplot(x=dataframe[column_name])
## plt.xlabel(" ")
```

Majority of the restaurant falls in dinning category

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c = "green", marker = "o" )
plt.xlabel("Type of Restaurant", c = "red", size = 20)
plt.ylabel("Votes", c = "blue", size = 20)
```

Dinning restaurants have received maximum votes

```python
dataframe.head()
```

```python
## Histogram

plt.hist(dataframe ['rate'],bins = 10)  ## bins are these 5 elevations
plt.title("Rating Distribution")
plt.show()
```

The majority restaurants received the ratings from 3.5 to 4


# Average order spending by couples


```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data, palette = 'hls')
```

```python
# conclusion - The majority of couples prefer restaurants with an approximate cost of 300 rupees.
```

Which mode receives maximum rating

```python
dataframe.head()
```

```python
# Box plot

plt.figure(figsize=(6,6))
sns.boxplot(x = 'online_order',y = 'rate', data = dataframe, palette  = 'hls')
```

Conclusion - offline order received lower rating in comparison to online order

```python
pivot_table = dataframe.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc = 'size', fill_value = 0)
sns.heatmap(pivot_table, annot = True, cmap = 'YlGnBu', fmt = 'd')
plt.title("HeatMap")
plt.xlabel("Online Order")
plt.ylabel("Listed In(Type)")
plt.show()
```

**CONCLUSION:Dining restaurants primarily accept offline orders, whereas cafes primarily receives online orders. This suggests that clients preferred orders in person at restaurants, but prefer online ordering at cafes**
